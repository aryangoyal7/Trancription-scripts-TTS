```python
f = open(f'{file_path}', 'r')
 #Let's take sets of 10 words and take it as our one training sample
total_time = 0
tstart = 0
file_number = 0
for labels in f:
        labels = labels.strip('\n')
        l = labels.split('\t')
        tmax = float(l[2]) * 1000
        tmin = float(l[1]) * 1000

        time = tmax - tmin

        total_time = total_time + time
        if( total_time >= 10000):
          file_name = 'wav/' + l[2] + '.wav'
          newAudio = AudioSegment.from_wav('/content/drive/MyDrive/Gradtts/sample1.wav')
          newAudio = audio[tstart: tstart + total_time]
          #newAudio.export(file_name, format='wav')
          newAudio.export(f'/content/drive/MyDrive/Gradtts/test1/{file_number}.wav', format='wav')
          print(f"Exported file {file_name} successfully")
            
          total_time = 0
          tstart = tmax 
          file_number += 1
          






        

        
```
